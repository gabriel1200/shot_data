```python

# Load the data
# Assume we have a DataFrame 'df' with columns: 'Season', 'Player', and various stats



import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

import plotly.colors as colors
def create_pretty_name(stat):
    pretty_names = {
        'PTS': 'Pts',
        'TS_PCT': 'TS%',
        'RA_FGA': 'Rim FGA',
        'RA_FG_PCT': 'Rim FG%',
        'DRIVES': 'Drives',
        'PULL_UP_FGA': 'Pull-up',
        'PULL_UP_EFG_PCT': 'Pull-up%',
        'FTA': 'FTAs',
        'POTENTIAL_AST': 'Pot. Ast',
        'TOUCHES': 'Touches',
        'CATCH_SHOOT_FGA': 'C&S',
        'CATCH_SHOOT_EFG_PCT': 'C&S eFG%'
    }
    return pretty_names.get(stat, stat.replace('_', ' ').title())

def create_pretty_name(stat):
    pretty_names = {
        'PTS': 'Points',
        'TS_PCT': 'TS%',
        'RA_FGA': 'Rim FGA',
        'RA_FG_PCT': 'Rim%',
        'DRIVES': 'Drives',
        'PULL_UP_FGA': 'Pull-up',
        'PULL_UP_EFG_PCT': 'PUeFG%',
        'FTA': 'FTAs',
        'POTENTIAL_AST': 'PotAst',
        'TOUCHES': 'Touches',
        'CATCH_SHOOT_FGA': 'C&S',
        'CATCH_SHOOT_EFG_PCT': 'C&S%'
    }
    return pretty_names.get(stat, stat.replace('_', ' ').title())

def create_career_graphic(player_id, player_data, stats_to_plot):
    # Ensure the data starts from 2014
    player_data = player_data[player_data['year'] >= 2014]
    
    # Calculate percentiles for each stat across all players per year
    for stat in stats_to_plot:
        if 'pct' not in stat.lower():
            player_data[stat]=75* player_data[stat]/player_data['POSS']
        player_data[f'{stat}_percentile'] = player_data.groupby('year')[stat].rank(pct=True)
    
    # Now filter by the selected player
    player_data = player_data[player_data['PLAYER_ID'] == player_id].reset_index(drop=True)
    
    years = player_data['year'].tolist()
    player_name = player_data['PLAYER_NAME'].iloc[0]
    
    # Create subplots with 'domain' type
    fig = make_subplots(
        rows=len(years),
        cols=len(stats_to_plot),
        specs=[[{'type': 'domain'} for _ in stats_to_plot] for _ in years],
        vertical_spacing=0.05,
        horizontal_spacing=0.02
    )
    
    # Get plasma colorscale
    plasma_colors = colors.sequential.Plasma

    for i, year in enumerate(years):
        for j, stat in enumerate(stats_to_plot):
            value = player_data[f'{stat}_percentile'].iloc[i]
            
            # Get color from plasma colorscale
            color = colors.sample_colorscale(plasma_colors, value)[0]
            
            # Create circular gauge
            fig.add_trace(
                go.Indicator(
                    mode="gauge+number",
                    value=round(value * 100,0),
                    domain={'row': i, 'column': j},
                    gauge={
                        'axis': {'range': [None, 100], 'visible': False},
                        'bar': {'color': color},
                        'bgcolor': "rgba(255,255,255,0.1)",
                        'borderwidth': 2,
                        'bordercolor': "#211a1d",
                        'shape': "angular",
                    },
                    number={'font': {'size': 16, 'color': 'white'}, 'suffix': "%"},
                ),
                row=i+1, col=j+1
            )
    
    # Add year annotations on the left side
    annotations = []
    carry = .006
    for i, year in enumerate(years):
        # The y position should align with the midpoint of each row
        annotations.append(
            dict(
                x=-0.06,  # Adjust the x position to be outside the plot
                y= 1-(i*1 / len(years)) - (carry*i),  # Adjust y position to align with the center of each row
                text=str(year),
                font=dict(size=24, color="white"),
                showarrow=False,
                xref="paper",
                yref="paper"
            )
        )
    
    # Add column names above each respective column
    carry=0.1
    for j, stat in enumerate(stats_to_plot):
        carry+=.045
        annotations.append(
            dict(
                x=(j + carry) / len(stats_to_plot),  # Center the text above each column
                y=1.03,  # Position above the plot
                text=create_pretty_name(stat),
                font=dict(size=20, color="white"),
                showarrow=False,
                xref="paper",
                yref="paper"
            )
        )
    
    # Update layout
    fig.update_layout(
        title=f"Career Percentile Rankings for {player_name}",
        title_x=.5,
        title_y=.99,
        font={'color': "white",'size':20},
        plot_bgcolor='#211a1d',
        paper_bgcolor='#211a1d',
        height=120 * len(years),
        width=120 * len(stats_to_plot),
        margin=dict(l=100, r=50, t=100, b=50),  # Increase left margin for year annotations
        annotations=annotations
    )
    
    return fig



# Example usage:
# Assuming you have player_data DataFrame and other necessary variables
# fig = create_career_graphic(player_id, player_data, stats_to_plot)
# fig.show()

# Example usage:
# Assuming you have player_data DataFrame and other necessary variables
# fig = create_career_graphic(player_id, player_data, stats_to_plot)
# fig.show()

# Example usage:
# Assuming you have player_data DataFrame and other necessary variables
# fig = create_career_graphic(player_id, player_data, stats_to_plot)
# fig.show()
# fig.show()
# fig = create_career_graphic(player_id, player_data, stats_to_plot)
# plt.show()
# fig = create_career_graphic(player_id, player_data, stats_to_plot)
# plt.show()
# Example usage
data = data[data['year'] >= 2014]
fig = create_career_graphic(203954,data,data_col)
fig.show()
```
