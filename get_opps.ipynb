```python
import pandas as pd
import os
from nba_api.stats.static import players,teams

nba_teams = teams.get_teams()
teamids= {}
team_acr = {}
team_ids = {
    'SEA': 1610612760,  # Seattle SuperSonics
    'CHH': 1610612766,  # Charlotte Hornets (original)
    'NJN': 1610612751,  # New Jersey Nets
    'NOH': 1610612740   # New Orleans Hornets (now New Orleans Pelicans)
}
team_extra={'1610612760': 'SEA', '1610612766': 'CHH', '1610612751': 'NJN', '1610612740': 'NOH'}

team_acr.update(team_ids)

teamids.update(team_extra)
opp_frames=[]
for org in nba_teams:
    teamids[str(org['id'])]=org['abbreviation']
    team_acr[org['abbreviation']] = str(org['id'])
for year in range(1997,2025):
    dir_path='team/'+str(year)+'ps/'
    csvs= os.listdir(dir_path)
    csvs =[csv for csv in csvs if '.csv' in csv]

    csvs =[csv for csv in csvs if 'vs' not in csv]
    csvs =[csv for csv in csvs if 'avg' not in csv]
    print(csvs)
    for csv in csvs:
        file_path = dir_path+csv
        team_id = csv.split('.')[0]

        df = pd.read_csv(file_path)
        df=df.sort_values(by='SHOT_ID')
        df=df[['VTM']]
        
        df =df.drop_duplicates()
        

        team_ac = teamids[team_id]
        #print(team_ac)
        df=df[df.VTM!=team_ac]
        df.columns=['OPP']

        df['OPP_ID']=df['OPP'].map(team_acr)
        df['OPP_ID'] = df['OPP_ID'].astype(str)
        df['TEAM_ID']= int(team_id)
        df['TEAM']= team_ac
        df['Round'] = [i+1 for i in range(0,len(df))]
        df['year']=year
        opp_frames.append(df)

opp_matchups = pd.concat(opp_frames)
opp_matchups.to_csv('opponentmatch.csv',index=False)
opp_matchups
```
